```python
# 라이브러리 불러오기
import pandas as pd
import seaborn as sns
```

```python
# titanic 데이터셋에서 age, sex등 5개 열을 선택하여 데이터프레임 만들기
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]
df.head()
```

```python
# groupby() 호출
grouped = df.groupby(['class'], observed=True)

# 그룹화된 데이터 확인
print(grouped)
```

```python
# 그룹 객체를 for문으로 출력: head() 메소드로 첫 5행만 출력
for key, group in grouped:
    print('* key :', key)
    print('* number:', len(group))
    print(group.head())
    print('\n')
```

```python
# 연산 메소드 적용
# 숫자형 데이터만 자동으로 계산함
average = grouped.mean(numeric_only=True)
average
```

```python
# 개별 그룹 필터링하기
group3 = grouped.get_group(('First',))
group3.head()
```

```python
# class 열, sex 열을 기준으로 분할
grouped_two = df.groupby(['class', 'sex'], observed=True)

# 특정 그룹만 출력하기
group_first_female = grouped_two.get_group(('First', 'female'))
group_first_female

# grouped_two 그룹 객체를 for문으로 출력
for key, group in grouped_two:
    print('* key :', key)
    print('* number:', len(group))
    print(group)
    print('\n')
```

```python
# grouped_two 그룹 객체에 연산 메소드 적용
average_two = grouped_two.mean(numeric_only=True)
average_two
```

```python
# grouped_two 그룹 객체에섯 개별 그룹 선택하기
group3f = grouped_two.get_group(('Third', 'female'))
group3f.head()
```
