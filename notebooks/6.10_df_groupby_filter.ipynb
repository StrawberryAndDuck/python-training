```python
# 라이브러리 불러오기
import pandas as pd
import seaborn as sns

# titanic 데이터셋에서 age, sex, class, fare, survived 5개 열을 선택해서 데이터프레임 만들기
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]

# class 열을 기준으로 분할
grouped = df.groupby(['class'], observed=True)
print(type(grouped))
print(grouped.size())

# 그룹별로 첫 2행 확인
grouped_head = grouped.head(2)

# 그룹별로 묶고 첫 2행 확인_인덱스 무
grouped_head = grouped.apply(lambda x: x.head(2)).reset_index(drop=True)
print(grouped_head)

# 그룹별로 묶고 첫 2행 확인_인덱스 유
grouped_head_2 = grouped.apply(lambda x: x.head(2))
print(grouped_head_2)
```

```python
# 각 그룹의 첫번째 데이터를 확인
grouped_first = grouped.nth(1)
print(grouped_first)

# 필터링이 적용된 그룹 객체의 열 부분 집합
grouped_col_first = grouped[["sex", "survived"]].nth(1)
print(grouped_col_first)

# 데이터 개수가 200개 이상인 그룹만을 필터링하여 데이터프레임으로 반환
grouped_filter = (grouped.filter(lambda x: len(x) >= 200))
print(grouped_filter)
print(grouped_filter['class'].value_counts())
```

```python
# age열의 평균이 30보다 작은 그룹만 필터링하여 데이터프레임으로 반환
print(grouped['age'].mean())
age_filter = grouped.filter(lambda x: x['age'].mean() < 30)
age_filter
```
