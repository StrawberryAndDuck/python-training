# 1. 데이터 로드

```python
import os
import kagglehub
import numpy as np
import pandas as pd
```

```python
# Download latest version
path = kagglehub.dataset_download("olistbr/marketing-funnel-olist")

print("Path to dataset files:", path)

csv_paths = []

# 2. List files in the downloaded directory (might be nested)
for root, dirs, files in os.walk(path):
    for file in files:
        if file.endswith(".csv"):
            csv_path = os.path.join(root, file)
            csv_paths.append(csv_path)

# 3. Load the CSV
dfs = [(csv_path, pd.read_csv(csv_path)) for csv_path in csv_paths]

# 4. Preview the data
for csv_path, df in dfs:
    print(f"Preview of {csv_path}:")
    display(df.head())
```

```python
df_mql = dfs[0][1]
df_closed_deal = dfs[1][1]
```

# 2. MQL(잠재 고객)으로부터 판매자(Seller)로의 전환율을 시각화

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# 한글 깨짐 및 마이너스 기호 설정
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

# 각 단계별 mql_id 고유 갯수 계산
mql_count = df_mql['mql_id'].nunique()
closed_deal_count = df_closed_deal['mql_id'].nunique()

# 퍼널 단계 데이터 구성
funnel_data = pd.DataFrame({
    'Stage': ['MQL', 'Closed Deal'],
    'Count': [mql_count, closed_deal_count]
})

max_count = funnel_data['Count'].max()

# 깔때기 모양(중앙 정렬된 수평 바) 그리기
fig, ax = plt.subplots(figsize=(8, 4))
for i, row in funnel_data.iterrows():
    stage = row['Stage']
    count = row['Count']
    left = (max_count - count) / 2   # 중앙 정렬을 위한 좌측 여백
    ax.barh(stage, count, left=left, height=0.5, color='skyblue')
    ax.text(max_count/2, i, f"{count}", ha='center', va='center', fontsize=12)

ax.set_xlim(0, max_count)
ax.set_xlabel("건수")
plt.title("퍼널 분석: MQL -> Closed Deal")
ax.invert_yaxis()  # y축 순서를 반전시켜 MQL이 위에 오도록 함
plt.show()
```

# 3. 이탈 및 전환된 MQL의 특성들을 EDA

```python
conversioned_index = df_mql.join(other=df_closed_deal.set_index('mql_id'), on='mql_id', how='inner').index
df_bounced_mql = df_mql.drop(index=conversioned_index)
df_bounced_mql
```

```python
df_bounced_mql.drop(columns=['mql_id', 'landing_page_id'], inplace=True)
```

```python
df_bounced_mql
```

```python
def visualize_distribution_by_origin(df: pd.DataFrame, origin_col: str, state: str):
    """
    Visualize the distribution of bounced MQLs by origin (내림차순 순서).
    
    Parameters:
    - df: DataFrame containing bounced MQLs
    - origin_col: Column name representing the origin of MQLs
    """
    plt.figure(figsize=(10, 5))
    
    # 내림차순으로 정렬된 순서 계산
    order = df[origin_col].value_counts().index
    sns.countplot(data=df, x=origin_col, palette='viridis', order=order)
    
    plt.title(f'{state} MQL Origin Distribution')
    plt.xlabel('Origin')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Visualize the distribution of bounced MQLs by origin using bar plot
visualize_distribution_by_origin(df=df_bounced_mql, origin_col='origin', state='Bounced')
```

- 이탈을 적게하는 매체는 주로 리퍼럴, 디스플레이 광고 등이 됨
- 주로 이탈이 발생하는 매체는 오가닉 콘텐츠, 페이드 마케팅, 소셜 미디어 접한 경우이다.
- 이 다음으로 해야할 것은 실제 전환된 고객(Seller)들에서도 이러한 가설을 검증하는 것이다.

```python
df_converisioned_mql = df_mql.loc[conversioned_index]
df_converisioned_mql
```

```python
visualize_distribution_by_origin(df=df_converisioned_mql, origin_col='origin', state='Conversioned')
```

- 의외의 결과가 나왔다. 오가닉 콘텐츠랑 페이드 마키팅에서 이탈률의 비율이 높아 반대로 전환된 고객은 해당 매체들에 대한 비율이 낮을 줄 알았으나,,
- 오히려 높은 이유가 해당 매체들로 접근하는 MQL들의 모수가 많다는 놀라운 인사이트를 얻었다.
- 고로, 다음으로 검증해야할 부분은 실제로 모든 MQL중에서 매체별 접근 비율을 시각해보는 것이다.

```python
def visualize_versus_distribution_by_origin(
    df_all: pd.DataFrame,
    df_conv: pd.DataFrame,
    df_bounced: pd.DataFrame,
    origin_col: str,
    state: str
):
    """
    전체 잠재 고객(df_all)에서 origin별 전환(df_conv) 및 이탈(df_bounced) 고객 분포를
    스택형 막대그래프로 시각화하고, 각 bar마다 전환/이탈 비율(백분율)을 표시한다.
    
    Parameters:
    - df_all: 전체 잠재 고객 DataFrame
    - df_conv: 전환된 고객 DataFrame
    - df_bounced: 이탈된 고객 DataFrame
    - origin_col: origin 컬럼 이름
    - state: 시각화 제목에 들어갈 텍스트
    """

    # 내림차순 순서 (전체 잠재고객 기준)
    order = df_all[origin_col].value_counts().index

    # origin별 count 계산 (전체, 전환, 이탈)
    conv_counts = df_conv.groupby(origin_col).size().reindex(order, fill_value=0)
    bounced_counts = df_bounced.groupby(origin_col).size().reindex(order, fill_value=0)

    x = np.arange(len(order))

    _, ax = plt.subplots(figsize=(10, 5))
    width = 0.6
    # 전환된 고객: 초록색 바
    ax.bar(x, conv_counts, width, label="전환", color="green")
    # 이탈된 고객: 빨간색 바 (전환 위에 스택)
    ax.bar(x, bounced_counts, width, bottom=conv_counts, label="이탈", color="red")

    # 각 bar에 전환/이탈 비율(백분율) 텍스트 추가
    for i, (conv, bounce) in enumerate(zip(conv_counts, bounced_counts)):
        total = conv + bounce
        if total > 0:
            conv_pct = conv / total * 100
            bounce_pct = bounce / total * 100

            # 전환된 고객 텍스트: 전환 바 중앙에 표시
            ax.text(x[i], conv/2, f"{conv_pct:.1f}%", ha='center', va='center', color='white', fontsize=11)
            # 이탈된 고객 텍스트: 이탈 바 중앙에 표시
            ax.text(x[i], conv + bounce/2, f"{bounce_pct:.1f}%", ha='center', va='center', color='white', fontsize=11)
        else:
            # 전체 0인 경우, 아무 텍스트도 표시하지 않음.
            pass

    ax.set_xticks(x)
    ax.set_xticklabels(order, rotation=45)
    ax.set_xlabel("Origin")
    ax.set_ylabel("건수")
    ax.set_title(f'{state} MQL Origin Distribution')
    # ax.legend()
    # plt.tight_layout()
    plt.show()
```

```python
visualize_versus_distribution_by_origin(
    df_all=df_mql,
    df_conv=df_converisioned_mql,
    df_bounced=df_bounced_mql,
    origin_col='origin',
    state='Conversioned vs Bounced'
)
```

- 최종적으로 얻은 인사이트는 백분률 기준 가장 효율이 좋은 매체는 unknown -> organic -> paid 순
