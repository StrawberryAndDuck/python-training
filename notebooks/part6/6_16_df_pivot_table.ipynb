```python
# 라이브러리 불러오기
import pandas as pd
import seaborn as sns

# titanic 데이터셋에서 5개 열 선택해서 데이터프레임 만들기
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]
df.head()
```

```python
# 평균 집계 - 행, 열, 값, 집계에 사용할 열 1개씩 지정
pdf = pd.pivot_table(df,
                    index='class',
                    columns='sex',
                    values='age',
                    aggfunc='mean',
                    observed=True)

pdf
```

```python
# 생존율, 생존자 수 집계
pdf2 = pd.pivot_table(df,
                    index='class',
                    columns='sex',
                    values='survived',
                    aggfunc=['mean', 'sum'],
                    observed=True)

pdf2
```

```python
# 평균 나이, 최대 요금 집계 - 행, 열, 값에 사용할 열을 2개 이상 지정 가능
pdf3 = pd.pivot_table(df,
                    index=['class', 'sex'],
                    columns='survived',
                    values=['age', 'fare'],
                    aggfunc=['mean', 'max', 'min'],
                    observed=True)

pdf3
```
