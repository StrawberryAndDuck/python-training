## 칼럼 이름 변경

```python
# 칼럼 이름 확인
display(dating_df.columns)

# 칼럼 갯수 확인
display(len(dating_df.columns))
```

```python
'string'.replace('s', 'a')
```

```python
new_columns = []

for i in dating_df.columns:
    if i.startswith('pref_o'):
        i = i.replace('pref_o', 'o_important')
    elif i.endswith('_o'):
        i = 'o_score_' + i.replace('_o','')
    elif i.endswith('_important'):
        i = 'i_important_' + i.replace('_important', '')
    elif i.endswith('_partner'):
        i = 'i_score_' + i.replace('_partner', '')
    new_columns.append(i)
```

```python
dating_df.columns = new_columns
```

```python
dating_df.head()
```

```python
# 상대방 age, race 정보 칼럼명 수정
dating_df = dating_df.rename({'o_score_race': 'race_o', 'o_score_age': 'age_o'}, axis = 1)
dating_df.head()
```

## 결측치 처리

```python
# 결측치 확인
dating_df.isna().mean()
```

```python
drop_columns = []

for i in dating_df.columns:
    if i.startswith('o_import'):
        drop_columns.append(i)
    elif i.startswith('i_important'):
        drop_columns.append(i)
```

```python
drop_columns
```

```python
dating_df = dating_df.dropna(subset = drop_columns)
dating_df.isna().mean() * 100
```

```python
# Nan값 처리
dating_df = dating_df.fillna(-99)
dating_df.head(20)
```
