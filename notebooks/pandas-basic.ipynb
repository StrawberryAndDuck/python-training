## 2-1. 시리즈(Series)

### 시리즈 만들기

```python
# 라이브러리 불러오기
import pandas as pd

# k:v 구조를 갖는 갖는 딕셔너리를 만들고, 변수 dict_data에 저장
dict_data = {'a': 1, 'b': 2, 'c': 3}

# Pandas Series 클래스 생성자를 사용해 딕셔러니(dict_data)를 시리즈로 변환, 변수 sr에 저장
sr = pd.Series(dict_data)

# 변수 sr의 자료형 출력
print(type(sr))

# 변수 sr에 저장되어 있는 시리즈 객체를 출력
print(sr)
print(sr.values)

```

### 인덱스 구조

```python
# list를 Series로 변환하여 변수 l에 저장
data = ['2019-01-01', 3.14, 'ABC', 100, True]
l = pd.Series(data)
print(l)

# 인덱스 배열 정보 확인
idx = l.index
print(idx) # RangeIndex(start=0, stop=5, step=1) start : 인덱스 시작값 / stop : 인덱스 끝나기 직전값 / step : d; 인덱스 값 간격

# 값 배열 변수 val에 저장
val = l.values
print(val)

# 시리즈 배열을 구성하는 원소의 자료형
print(l.dtype)

# 시리즈 배열의 크기, 원소의 개수 출력
print(len(l))

# 시리즈 배열의 형태
print(l.shape) # (행수, 열수)

# 시리즈 배열의 차원
print(l.ndim)
```

### 원소 선택

```python
# tuple을 Series로 변환(index 옵션 지정)
tup_data = ('영인', '2010-05-01', '여', True)
sr = pd.Series(tup_data, index=['이름', '생년월일', '성별', '학생여부'])
sr_2 = pd.Series(tup_data, ) 

print(sr) # 인덱스값 출력
print(sr_2) # 인덱스 번호 출력

# 원소 1개 선택
print(sr[0]) # 정수형 위치 인덱스
print(sr['이름']) # 인덱스 이름

# 여러 개의 원소를 선택(인덱스 리스트 활용)
print(sr[[1, 2]])
print('\n')
print(sr[['생년월일', '성별']])

# 여러 개의 원소를 선택(인덱스 범위 지정)
print(sr[1:2])
print('\n')
print(sr['생년월일' : '성별'])
```

### 시리즈 객체를 만드는 특수한 방법

```python
# 빈 Series 객체 생성
pd.Series()

# Series 원소 선택
sr = pd.Series(5)
print(sr)

# index 배열 갯수 만큼 원소값 복제
pd.Series(5, index=["a", "b", "c"])
a = pd.Series("한", index=["a", "b", "c"])
print(a)



```

## 2-2. 데이터프레임

### 데이터프레임 만들기

```python
# column name을 key로 하고, list를 value로 갖는 dictionary 정의 (2 dimention array) 
dict_data = {'c0':[1, 2, 3], 'c1':[4, 5, 6], 'c2':[7, 8, 9], 'c3':[10, 11, 12], 'c4':[13, 14, 15]}

# 판다스 데이터프레임 객체로 dictionary -> Dataframe으로 변환 후 변수 df에 저장
df = pd.DataFrame(dict_data)

# df 변수에 할당된 데이터프레임 객체 출력
print(df)
# 데이터프레임 type 출력
print(type(df))
```

### 행 인덱스/열 이름 설정

```python
# 행 이덱스/ 열 이름 지정하여 데이터프레임 만들기
df = pd.DataFrame([[15, '남', '덕영중'], [17, '여', '수리중']], # 리스트처럼 순서가 있고, 반복 가능한 자료형 가능
                index = ['준서', '예은'],
                columns = ['나이', '성별', '학교'])

# 데이터프레임 출력하기
print(df)

# 행 인덱스 확인하기
print(df.index)

# 열 인덱스 확인하기
print(df.columns)

# 행 인덱스, 열 이름 변경하기
df.index = ['학생1', '학생2']
df.columns = ['연령', '남녀', '소속']

print(df)
print('\n')
print(df.index)
print('\n')
print(df.columns)

# 열 이름 중, '나이' -> '연령', '성별' -> '남녀', '학교' -> 소속으로 변경
df = df.rename(columns={'나이':'연령', '성별':'남녀', '학교':'소속'})
df = df.rename(index={'학생1':'학생2', '학생2':'학생3'})
print(df)
```
