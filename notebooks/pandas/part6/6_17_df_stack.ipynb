```python
# 라이브러리 불러오기
import pandas as pd
import seaborn as sns

# titanic 데이터셋에서 5개 열 선택하여 데이터프레임 생성
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]

# 피벗 테이블로 멀티 인덱스를 갖틑 데이터프레임 만들기
df2 = pd.pivot_table(df,
                    index=['class', 'survived'],
                    columns='sex',
                    values='age',
                    aggfunc='mean',
                    observed=True)
df2
```

```python
# stack 적용 - 열을 행으로 바꾼다
df_stacked = df2.stack()
df_stacked
```

```python
# unstack 적용 - 행에 있는 데이터 열로 이동
df_unstacked = df2.unstack()
print(df_unstacked)

# unstack 적용 - level=0 지정
df_unstacked2 = df2.unstack(level=0)
print(df_unstacked2)

# unstack 적용 - level=1 지정
df_stacked2 = df_unstacked2.stack(level=1, future_stack=True)
print(df_stacked2)
```
